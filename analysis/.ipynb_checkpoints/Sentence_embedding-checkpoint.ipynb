```python
from transformers import pipeline, AutoTokenizer, AutoModel, AutoModelForSequenceClassification
import numpy as np
import pandas as pd
import torch
from sklearn.metrics.pairwise import cosine_similarity
from scipy.spatial.distance import cosine
from sentence_transformers import SentenceTransformer, util, models
```

```python
model_name = 'climatebert/distilroberta-base-climate-f'

tokenizer = AutoTokenizer.from_pretrained(model_name)

model = AutoModel.from_pretrained(model_name, output_hidden_states = True)
```

```python
z = ["This product is environmentally friendly.",
    'The offering happens to be good for nature.',
    'The product is unsustainable and destroys the environment.']
```

```python
#nitialize our model and tokenizer:
# tokenizer = AutoTokenizer.from_pretrained('sentence-transformers/bert-base-nli-mean-tokens')
# model = AutoModel.from_pretrained('sentence-transformers/bert-base-nli-mean-tokens')
# # ##Tokenize the sentences like before:
# sent = [
#     "Three years later, the coffin was still full of Jello.",
#     "The fish dreamed of escaping the fishbowl and into the toilet where he saw his friend go.",
#     "The person box was packed with jelly many dozens of months later.",
#     "He found a leprechaun in his walnut shell."
# ]
# initialize dictionary: stores tokenized sentences
token = {'input_ids': [], 'attention_mask': []}
for sentence in z:
    # encode each sentence, append to dictionary
    new_token = tokenizer.encode_plus(sentence, max_length=128,
                                       truncation=True, padding='max_length',
                                       return_tensors='pt')
    token['input_ids'].append(new_token['input_ids'][0])
    token['attention_mask'].append(new_token['attention_mask'][0])
# reformat list of tensors to single tensor
token['input_ids'] = torch.stack(token['input_ids'])
token['attention_mask'] = torch.stack(token['attention_mask'])
```

```python
output = model(**token)
embeddings = output.last_hidden_state
```

```python
embeddings_test = output.last_hidden_state[:,0,:]
```

```python
att_mask = token['attention_mask']
mask = att_mask.unsqueeze(-1).expand(embeddings.size()).float()
mask_embeddings = embeddings * mask

summed = torch.sum(mask_embeddings, 1)
summed_mask = torch.clamp(mask.sum(1), min=1e-9)
mean_pooled = summed / summed_mask
mean_pooled
```

```python
#Let's calculate cosine similarity for sentence 0:
# convert from PyTorch tensor to numpy array
mean_pooled = mean_pooled.detach().numpy()
# calculate
cosine_similarity(
    [mean_pooled[0]],
    mean_pooled[1:]
)
```

```python
embeddings_test = embeddings_test.detach().numpy()

cosine_similarity(
    [embeddings_test[0]],
    embeddings_test[1:]
)
```

```python
encoded_inputs = tokenizer(z)

storage = []
for i, text in enumerate(encoded_inputs['input_ids']):
    
    tokens_tensor = torch.tensor([encoded_inputs['input_ids'][i]])
    segments_tensors = torch.tensor([encoded_inputs['attention_mask'][i]])
    print (tokens_tensor)
    print (segments_tensors)

    # Run the text through BERT, and collect all of the hidden states produced
    # from all 12 layers. 
    with torch.no_grad():

        outputs = model(tokens_tensor, segments_tensors)

        # Evaluating the model will return a different number of objects based on 
        # how it's  configured in the `from_pretrained` call earlier. In this case, 
        # becase we set `output_hidden_states = True`, the third item will be the 
        # hidden states from all layers. See the documentation for more details:
        # https://huggingface.co/transformers/model_doc/bert.html#bertmodel
        hidden_states = outputs[2]


    # `hidden_states` has shape [13 x 1 x 22 x 768]

    # `token_vecs` is a tensor with shape [22 x 768]
    token_vecs = hidden_states[-2][0]

    # Calculate the average of all 22 token vectors.
    sentence_embedding = torch.mean(token_vecs, dim=0)
    print (sentence_embedding[:10])
    storage.append((text,sentence_embedding))
```

```python
def plot_clusters(df):
    embeddings = list(df['new_embeddings'])
    embeddings = np.array(embeddings)
    
    min_clusters = 2
    max_clusters = 20
    inertias = []
    for k in range(min_clusters, max_clusters):
        kmeans = KMeans(n_clusters = k, init = 'k-means++',random_state = 1542, n_init = 10).fit(embeddings)
        inertias.append(kmeans.inertia_)
    
    plt.plot(range(min_clusters, max_clusters), inertias, '-o')
    plt.xlabel('Number of clusters (k)')
    plt.ylabel('Inertia')
    plt.title('Scree plot')
    return plt.show()
```

```python
def add_clusters(df,k):
    embeddings = list(df['new_embeddings'])
    embeddings = np.array(embeddings)
    kmeans = KMeans(n_clusters = k, init = 'k-means++',random_state = 1542, n_init = 10).fit(embeddings)
    clusters = kmeans.labels_
    df['clusters'] = clusters
    return df
```

```python
def plot_wordcloud(df, cluster):
    cluster_text = ' '.join([sentence for sentence in df[df['clusters'] == cluster]['sentence'].str.lower()])
    word_cloud = WordCloud(collocation_threshold = 2, width = 1000, height = 500, background_color = 'white'
                      ).generate(cluster_text)

    plt.figure(figsize = (10,5))
    plt.imshow(word_cloud)
    plt.axis('off')
    return plt.show
```

# Calculating Greenwashing Score - Sentiment ratio's 

```python
df_comb = pd.read_pickle('data_structured/comb.pkl')
```

```python
def count_sentiment(df, cluster, firm, doc_type ,sentiment):
    try:
        return df[(df['clusters']==cluster) & (df['company'] == firm) & (df['doc_type'] == doc_type)]['sentiment'].value_counts()[sentiment]
    except KeyError:
        return 0
    
```

An important question is - how do we handle 0 cases, as in when there are no negative or positive sentiment cases in the database. For example, let's say a company has no positive news in the database and some negative news - is the positive no negative sentiment ratio 0? Yes, since we subtract this number from the positive/negative ratio from the report - the discrepancy is technically maxed out in that case (i.e. there are no positive mentions in the articles about this particular area of the company). 
 - What if the positive and articles are both 0, but there are is some sort of pos/neg ratio in the report? Do I want to handle this as high discrepancy cases? Or should I just flag this as a 'no coverage' case? I think either way could work, but it might be smarter to flag these cases as 'no coverage' for now. So define a separate greenwash score function 
 - What if the positive articles are higher than 0, while negative equals 0 and there is a high pos/neg ratio in the report? I want there to be low discrepancy - so 0? Right now I'm saying the pos-neg-news ratio should be 1, but this may not work that well. Maybe flag it 'only positive coverage'?
 

```python
def ratio_articles(p, n):
    if (n == 0) & (p!=0):
        return 'positive coverage only'
    #1
    elif (n == 0) & (p == 0):
        return 'no coverage'
    #0
    elif (n!=0) & (p==0):
        return 'negative coverage only'
    else:
        return p/n
```

```python
def ratio_reports(p, n):
    if (n == 0) & (p!=0):
        return 'positive communication only'
    #1
    elif (n == 0) & (p == 0):
        return 'no communication'
    #0
    elif (n!=0) & (p==0):
        return 'negative communication only'
    else:
        return p/n
```

```python
#creating an answer generating data frame for the different cases of greenwashing:
data = {'negative communication only':['High transparency', 'Under-reporting (very) positive', 0],
         'positive communication only':['No evidence', 'Low discrepancy', '(Very) high discrepancy'],
         'no communication': ['Not applicable', 'Under-reporting positive', 'Under-reporting negative']}

answer_matrix = pd.DataFrame(data, index = ['no coverage',
                                            'positive coverage only',
                                            'negative coverage only'])

answer_matrix
```

Now I'm going to create a function from the above

```python
def gw_score(posneg_report, posneg_news):
    if isinstance(posneg_news, str) & isinstance(posneg_report, str):
        return (answer_matrix.loc[posneg_news, posneg_report])
    elif posneg_report == 'negative communication only':
        return (-posneg_news)
    elif (posneg_report == 'positive communication only'):
        if (posneg_news > 1):
            return ('Low discrepancy')
        else:
            return ('High discrepancy')
    elif (posneg_report == 'no communication'):
        if (posneg_news > 1):
            return ('Under-reporting positive')
        else:
            return ('Under-reporting negative')
    elif (posneg_report > 0):
        if posneg_news == 'no coverage':
            return ('No evidence')
        elif posneg_news == 'positive coverage only':
            return ('Low discrepancy')
        elif posneg_news == 'negative coverage only':
            return (posneg_report)
        else: 
            return (posneg_report - posneg_news)
```

```python
#light testing - calculate the difference for coca cola in cluster 5 for example
# I should probably switch this around, since I'm more interested in the greenwashing score per firm - for now will just do the average instead of the weighted average - can add that later 
gw_database = {}
for cluster in list(set(df_comb['clusters'])):
    firm_cluster_score = {}
    for firm in set(df_comb['company']):
        n_news_pos = count_sentiment(df_comb, cluster, firm, 'news', 'POSITIVE')
        n_news_neg = count_sentiment(df_comb, cluster, firm, 'news', 'NEGATIVE')
        n_report_pos = count_sentiment(df_comb, cluster, firm, 'report', 'POSITIVE')
        n_report_neg = count_sentiment(df_comb, cluster, firm, 'report', 'NEGATIVE')
        posneg_news = ratio_articles(n_news_pos,n_news_neg)
        posneg_report = ratio_reports(n_report_pos,n_report_neg)
        firm_cluster_score[firm] = gw_score(posneg_report, posneg_news)
    gw_database[cluster]= firm_cluster_score

    
gw_database
```
